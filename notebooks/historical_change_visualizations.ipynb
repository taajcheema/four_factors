```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import sklearn

from sklearn import linear_model

df = pd.read_excel("per_100_possessions.xlsx")
df.head()
```

```python
# Effective Field Goal % 

plt.figure(figsize=(20,9))
sns.lineplot(data=df, x='Season',y='eFG%')
plt.suptitle('NBA Effective Field Goal % 2000 - 2020', 
              fontsize=16, fontweight='bold')

plt.savefig('eFG%.png')
```

```python
# Turnover % 

plt.figure(figsize=(20,9))
sns.lineplot(data=df, x='Season',y='TOV%')
plt.suptitle('NBA Turnover % 2000 - 2020', 
              fontsize=16, fontweight='bold')

plt.savefig('turnovers.png')
```

```python
# Offensive Rebound % 

plt.figure(figsize=(20,9))
sns.lineplot(data=df, x='Season',y='ORB%')
plt.suptitle('NBA Offensive Rebound % 2000 - 2020', 
              fontsize=16, fontweight='bold')
plt.savefig('OREB%.png')
```

```python
# Free Throw Factor 

plt.figure(figsize=(20,9))
sns.lineplot(data=df, x='Season',y='FT/FGA')
plt.suptitle('NBA Free Throw Factor 2000 - 2020', 
              fontsize=16, fontweight='bold')
plt.savefig('FTF.png')
```
